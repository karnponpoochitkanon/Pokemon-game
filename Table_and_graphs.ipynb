```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# โหลดข้อมูล
df = pd.read_csv("game_stats.csv")

# กำหนดสไตล์สวย ๆ
sns.set(style="darkgrid", palette="pastel")

# ขนาดกราฟ
plt.figure(figsize=(10, 6))

# วาด histogram + KDE
sns.histplot(data=df, x="Distance", bins=20, kde=True, color="#69b3a2", edgecolor="black")

# เพิ่มชื่อกราฟ
plt.title("Distribution of Player Movement Distance", fontsize=16)
plt.xlabel("Distance Walked", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

# แสดงกราฟ
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("game_stats.csv")
df.columns = df.columns.str.strip()

plt.figure(figsize=(10, 6))
sns.histplot(df["YIM Battles"], bins=range(1, df["YIM Battles"].max() + 2),
             color="salmon", edgecolor="black")

plt.title("Distribution of YIM Battle Counts", fontsize=16)
plt.xlabel("Number of YIM Battles", fontsize=12)
plt.ylabel("Number of Players", fontsize=12)
plt.xticks(range(1, df["YIM Battles"].max() + 1))
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# โหลดข้อมูล
df = pd.read_csv("game_stats.csv")
df.columns = df.columns.str.strip()

# นับจำนวนตัวละคร
character_counts = df["Character"].value_counts()
labels = character_counts.index
sizes = character_counts.values

# สีเฉพาะ 3 ตัว
colors = ['lightcoral', 'turquoise', 'gold']

# สร้างกราฟ
fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts = ax.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors
)

# Legend ด้านบน
ax.legend(wedges, labels, title="Characters", loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=3)

# ชื่อกราฟ
ax.set_title("Most Frequently Selected Characters", fontsize=16)

plt.tight_layout()
plt.show()
```

```python
# โหลดข้อมูล
df = pd.read_csv("game_stats.csv")
df.columns = df.columns.str.strip()

# สร้างตารางความถี่
pokemon_counts = df["Total Pokemon"].value_counts().sort_index()

# วาดกราฟแท่ง
plt.figure(figsize=(10, 6))
sns.barplot(x=pokemon_counts.index, y=pokemon_counts.values, color="skyblue", edgecolor="black")

# ตกแต่งกราฟ
plt.title("Distribution of Total Pokemon per Player", fontsize=16)
plt.xlabel("Total Pokemon", fontsize=12)
plt.ylabel("Number of Players", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# โหลดข้อมูล
df = pd.read_csv("game_stats.csv")
df.columns = df.columns.str.strip()

# สร้างตารางความถี่ของจำนวนรอบการใช้ Heal Tree
heal_counts = df["Heal Tree Uses"].value_counts().sort_index()

# วาดกราฟแท่ง
plt.figure(figsize=(10, 6))
sns.barplot(x=heal_counts.index, y=heal_counts.values, color="mediumseagreen", edgecolor="black")

# ตกแต่งกราฟ
plt.title("Distribution of Heal Tree Usage Counts", fontsize=16)
plt.xlabel("Heal Tree Uses", fontsize=12)
plt.ylabel("Number of Players", fontsize=12)
plt.tight_layout()
plt.show()
```
