```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from IPython.display import display

df = pd.read_csv("game_stats.csv")
df.columns = df.columns.str.strip()

columns = ["Time(s)", "YIM Battles", "Distance", "Total Pokemon", "Heal Tree Uses"]
summary = df[columns].describe().T
summary["mode"] = df[columns].mode().iloc[0]

display(summary)
```

```python
df = pd.read_csv("game_stats.csv")

sns.set(style="darkgrid", palette="pastel")

plt.figure(figsize=(10, 6))

sns.histplot(data=df, x="Distance", bins=20, kde=True, color="#69b3a2", edgecolor="black")

plt.title("Distribution of Player Movement Distance", fontsize=16)
plt.xlabel("Distance Walked", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("game_stats.csv")
df.columns = df.columns.str.strip()

plt.figure(figsize=(10, 6))
plt.plot(df["Time(s)"], marker='', linestyle='-', color="steelblue")

plt.title("Play Time Line Chart", fontsize=16)
plt.xlabel("Player Index", fontsize=12)
plt.ylabel("Time(s)", fontsize=12)
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python

df = pd.read_csv("game_stats.csv")
df.columns = df.columns.str.strip()

plt.figure(figsize=(10, 6))
sns.histplot(df["YIM Battles"], bins=range(1, df["YIM Battles"].max() + 2),
             color="salmon", edgecolor="black")

plt.title("Distribution of YIM Battle Counts", fontsize=16)
plt.xlabel("Number of YIM Battles", fontsize=12)
plt.ylabel("Number of Players", fontsize=12)
plt.xticks(range(1, df["YIM Battles"].max() + 1))
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("game_stats.csv")
df.columns = df.columns.str.strip()


character_counts = df["Character"].value_counts()
labels = character_counts.index
sizes = character_counts.values

colors = ['lightcoral', 'turquoise', 'gold']

fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts = ax.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors
)


ax.legend(wedges, labels, title="Characters", loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=3)


ax.set_title("Most Frequently Selected Characters", fontsize=16)

plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("game_stats.csv")
df.columns = df.columns.str.strip()

pokemon_counts = df["Total Pokemon"].value_counts().sort_index()

plt.figure(figsize=(10, 6))
sns.barplot(x=pokemon_counts.index, y=pokemon_counts.values, color="skyblue", edgecolor="black")

plt.title("Distribution of Total Pokemon per Player", fontsize=16)
plt.xlabel("Total Pokemon", fontsize=12)
plt.ylabel("Number of Players", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("game_stats.csv")
df.columns = df.columns.str.strip()

heal_counts = df["Heal Tree Uses"].value_counts().sort_index()

plt.figure(figsize=(10, 6))
sns.barplot(x=heal_counts.index, y=heal_counts.values, color="mediumseagreen", edgecolor="black")

plt.title("Distribution of Heal Tree Usage Counts", fontsize=16)
plt.xlabel("Heal Tree Uses", fontsize=12)
plt.ylabel("Number of Players", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("game_stats.csv")
df.columns = df.columns.str.strip()

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="Total Pokemon", y="Time(s)", color="#5B9BD5", s=80, edgecolor="black")

plt.title("Relationship Between Total Pokemon and Play Time", fontsize=14)
plt.xlabel("Total Pokemon")
plt.ylabel("Time Played (seconds)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="Total Pokemon", y="Distance", color="#F4B183", s=80, edgecolor="black")
plt.title("Total Pokemon vs Distance Walked", fontsize=14)
plt.xlabel("Total Pokemon")
plt.ylabel("Distance Walked")
plt.grid(True)
plt.tight_layout()
plt.show()
```
